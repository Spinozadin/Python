```python
!pip install datasets
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

---

## ðŸŸ© Sort by Salary (1.25.1)

```python
df.sort_values(by='salary_year_avg', ascending=True, inplace=True)
df.head()
```

---

## ðŸŸ© Create Remote Column (1.25.2)

```python
df['is_remote'] = df['job_work_from_home']
```

---

## ðŸŸ© Extract Year from Date (1.25.3)

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_posted_year'] = df['job_posted_date'].dt.year
df.head(5)
```

---

## ðŸŸ¨ Drop Missing Salaries (1.25.4)

```python
df.dropna(subset=['salary_year_avg'], inplace=True)
```
